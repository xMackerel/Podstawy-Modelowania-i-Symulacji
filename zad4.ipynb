```python
import math
# e ^ -x/CC * R C[1] - (CC CC * R * omega * cos[x * omega] - sin[x * omega])/ (1 + CC^2 * R^2 * omega^2)

def solve(x):
  wynik = (CC * (CC * R * omega * math.cos(x * omega) - math.sin(x * omega)))/ (1 + CC**2 * R**2 * omega**2)
  return wynik
omega = 0.1
e = math.e
CC = 1
R = 1
lista = []
tmax = 2 * math.pi / omega
n = 100
i = 1
while(i <= 100):
    x = i + tmax/n
    lista.append(solve(x))
    i+=1
print(max(lista))
```
